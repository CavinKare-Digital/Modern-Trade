```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
#reading file in xlsx format
df = pd.read_excel(r'D:\Parthipan\CK\MT\Analytics Transfer\Big Bazaar.xlsx',sheet_name='Food SKU wise Sales Concept')
```

```python
df1 = df[df['Concept']=='BB'].reset_index()
df1.columns
#only for xlsb files
df1 = df1[pd.notnull(df1['Month'])]
df1 = df1[pd.notnull(df1['SKU Code'])]
df1.head()
```

```python
df1["Chain"] = "Big Bazaar"
```

```python
df1['SKU Code'] = df1['SKU Code'].astype(int).astype(str)
```

```python
df1['Date'] = '01' + '-' + df1['Month'].str[0:3]+ '-' + df1['Year'].map(str)
df1['Date'] = df1['Date'].str[:-2]
df1['Date'] = df1['Date'].apply(lambda x: dt.datetime.strptime(x,'%d-%b-%Y').date())
```

```python
df1['SKU Description']=df1['SKU Description'].str.strip()
```

```python
df1['SKU Code'] = df1['SKU Code'].astype(object)
```

```python
df1 = df1.rename(columns={"Store Code":"Store_Code_External", "SKU Code": "Item_Code_External", "SKU Description": "Item_Description_External", "Net Sales Units": "Sales_Qty", "Net Sales": "Sales_Value"})

```

```python
df1['Chain_Item_Code_External']=df1['Chain']+'_'+df1['Item_Code_External'].astype(int).astype(str)
```

```python
df1.sample(10)
```

```python
df1['Store_Code_External'] = df1['Store_Code_External'].astype(int).astype(str)
df1['Chain_Store_Code'] = df1['Chain']+'_'+df1['Store_Code_External'].astype(str)
```

```python
df1 = df1[['Chain', 'Date', 'Store_Code_External','Chain_Store_Code','Item_Code_External','Chain_Item_Code_External','Item_Description_External', 'Sales_Qty', 'Sales_Value']]
```
